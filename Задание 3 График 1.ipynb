```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 3))
x = np.arange(-10, 10.01, 0.2)
t = np.arange(-10, 11, 1)
plt.plot(x, np.arcsin((2 * abs(x)) / (1 + x ** 2)), 'r', t, np.arcsin((2 * abs(t)) / (1 + t ** 2)), 'yo', linewidth=2, markersize=8)
plt.show()
```
