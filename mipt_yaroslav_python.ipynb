```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 5))
x = np.arange(-6, 6.01, 0.01)
plt.plot(x, 1.1**(-2 * x) * np.sin(x) ** 2)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
x = np.arange(-6, 6.01, 0.01)
plt.plot(x, np.arcsin((2 * abs(x)) / (1 + x ** 2)))
plt.show()
```
